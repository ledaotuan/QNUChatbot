```python
import os
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
def get_javascript_doPostback(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        container = soup.find("section", class_="topic-content").find("div", class_="container").find("div", "PageColumns")  # Find the <div> with class "container"
        if container:
            link_elements = container.find_all("a", href=True)  # Find <a> elements within the container
            list_doPostback = [link["href"] for link in link_elements]   
            return list_doPostback[:-1] 
        else:
            print("list_doPostback: None")
            return []
    else:
        print(f"Request failed with status code: {response.status_code}")
        return []


def get_page_links(doPostback):
    global driver
    driver.execute_script(doPostback)
    # Wait for dynamic content to load
    wait = WebDriverWait(driver, 30)
    dynamic_element = wait.until(EC.presence_of_element_located((By.ID, "page")))
    dynamic_element_html = dynamic_element.get_attribute("outerHTML")
    soup = BeautifulSoup(dynamic_element_html, "html.parser")

    container = soup.find("section", class_="topic-content").find("div", class_="container")  # Find the <div> with class "container"
    if container:
        link_elements = container.find_all("a", href=True)  # Find <a> elements within the container
        links = []
        for link in link_elements:
            if (link["href"] != None) and (not link["href"] in links) and (not link["href"].startswith("javascript")):
                links.append(link["href"])
        return links
    else:
        print("Container: None")
        return []


def extract_info_from_link(link):
    print("Extract link: ", link)
    response = requests.get(link)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        content_element = soup.find("div", class_="tuyensinh")  # Update class name
    
        if content_element:
            extracted_text = content_element.get_text()
            file_name = link.split("/")[-1] + ".txt"
            with open("raw_data/"+file_name, "w", encoding="utf-8") as file:
                file.write(extracted_text)
            print("Data has been saved to", file_name)
        else:
            print("Content element not found.")
    else:
        print(f"Request failed with status code: {response.status_code}")
        
main_url = "https://tuyensinh.qnu.edu.vn/vi/tong-hop-nganh-dao-tao"
driver = webdriver.Firefox()
driver.get(main_url)
list_doPostback = get_javascript_doPostback(main_url)
for doPostback in list_doPostback:
    links = get_page_links(doPostback)
    
    for link in links:
        full_link = "https://tuyensinh.qnu.edu.vn/" + link
        extract_info_from_link(full_link)
driver.quit()
```
